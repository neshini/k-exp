```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
file = 'FinalPhotUV.dat'
df= pd.read_csv(file, sep='\s+')
df
```

```python
UVW1 = df.loc[:,'UVW1']
UVM2 = df.loc[:, 'UVM2']
B = df.loc[:, 'B']
V = df.loc[:,'V']

errUVW1 = df.loc[:,'errUVW1']
errUVM2 = df.loc[:, 'errUVM2']
errB = df.loc[:, 'errB']
errV = df.loc[:,'errV']



#for fig 3a
x1 = UVW1-V
y1 = B-V
x1err = abs(errUVW1-errV)
y1err = abs(errB-errV)
#for fig 3b
x2 = UVW1-UVM2
y2 = B-V
x2err = abs(errUVW1-errUVM2)
y2err = abs(errB-errV)
print(x1err)
```
